# Projeto: Análise de campanha de marketing bancária

#### Este projeto desenvolve uma análise a respeito da campanha de marketing retirada do dataset <a href = "https://archive.ics.uci.edu/dataset/222/bank+marketing">Bank Marketing</a> da UCI Machine Learning Repository. Essa campanha foi realizada por telefonemas por um banco português.

A variável alvo indica se o cliente aderiu (ou não) à oferta de um depósito a prazo (term deposit), que foi promovido durante uma campanha de marketing telefônico.

Um depósito a prazo  é um tipo de investimento seguro e de baixo risco, onde o cliente deixa seu dinheiro em um banco por um período determinado em troca de um rendimento fixo ou variável.

O cliente não pode sacar o dinheiro antes do prazo sem perder parte dos rendimentos (ou pagar uma penalidade). O banco oferece um juros fixo ou pré-definido no momento da aplicação. Em muitos países, esses depósitos são garantidos por instituições financeiras (como o FGC no Brasil, até certos valores). Diferente de uma conta poupança, não é possível movimentar o valor enquanto ele estiver aplicado.

- Identificar padrões comportamentais que influenciam a resposta positiva dos clientes.

- Visualizar os dados por meio de gráficos informativos, para facilitar a comunicação dos insights.

- Levantar hipóteses e propor possíveis estratégias com base nas descobertas da análise.

# Informações sobre o dataset

O dataset Bank Marketing foi originalmente disponibilizado por Paulo Cortez (Universidade do Minho) e Sérgio Moro (ISCTE-IUL) em 2012, com base em campanhas de marketing realizadas por um banco português. Seu objetivo principal é analisar campanhas de marketing direto por telefone com o intuito de prever se um cliente irá aderir a um depósito a prazo (a variável alvo y).

As campanhas envolviam ligações telefônicas, podendo haver múltiplos contatos com o mesmo cliente. O conjunto de dados completo (bank-full.csv) contém 45.211 registros, coletados entre maio de 2008 e novembro de 2010. Também está disponível uma amostra reduzida (bank.csv) com 4.521 registros, útil para testes com algoritmos mais exigentes computacionalmente.

O conjunto de dados é composto por 16 variáveis de entrada e 1 variável de saída, abrangendo informações como:

- Dados demográficos: idade, profissão, estado civil, escolaridade.

- Situação financeira: saldo médio anual, existência de empréstimos (pessoal ou habitacional), histórico de crédito.

- Detalhes da campanha atual: tipo de contato, duração da última ligação, mês e dia do último contato.

- Histórico de campanhas anteriores: número de contatos prévios e resultado dessas interações.

# Análise Exploratória dos Dados

##### O código abaixo serve para automatizar o grid dos gráficos, para que possamos analisar as variáveis do nosso dataset

```python
def plot_grid(df, cols, plot_type='countplot', target=None, n_cols=3):
    n_rows = -(-len(cols) // n_cols) 
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(n_cols*6, n_rows*4))
    axes = axes.flatten()

    for i, col in enumerate(cols):
        ax = axes[i]

        if plot_type == 'countplot':
            sns.countplot(data=df, x=col, order=df[col].value_counts().index, ax=ax)
            ax.set_title(f'Distribuição de {col}')
            ax.tick_params(axis='x', rotation=45)

        elif plot_type == 'stacked_bar' and target:
            crosstab = pd.crosstab(df[col], df[target], normalize='index')
            crosstab.plot(kind='bar', stacked=True, ax=ax, legend=False)
            ax.set_title(f'{col} vs {target}')
            ax.set_ylabel('Proporção')
            ax.tick_params(axis='x', rotation=45)

        elif plot_type == 'boxplot' and target:
            sns.boxplot(data=df, x=target, y=col, ax=ax)
            ax.set_title(f'{col} vs {target}')

        elif plot_type == 'barplot' and target:
            conv_rates = df.groupby(col)[target].mean().sort_values(ascending=False)
            sns.barplot(x=conv_rates.index, y=conv_rates.values, ax=ax)
            ax.set_title(f'Taxa de Conversão por {col}')
            ax.set_ylabel('Taxa de Conversão')
            ax.set_ylim(0, 1)
            ax.tick_params(axis='x', rotation=45)

        else:
            ax.axis('off')

    for j in range(i+1, len(axes)):
        fig.delaxes(axes[j])

    plt.tight_layout()
    plt.show()
```

##### Gráfico de distribuição das variáveis categóricas

```python
cat_cols = df_bank_full.select_dtypes(include='object').columns
plot_grid(df_bank_full, cat_cols, plot_type='countplot')
```

#####  Boxplots para variáveis numéricas

```python
import matplotlib.pyplot as plt
import seaborn as sns
import math

def plot_grid_dinamico(df, cols, plot_type, n_cols=3, target=None):

    n_plots = len(cols)
        
    n_rows = math.ceil(n_plots / n_cols)
    
    fig_width = n_cols * 5
    fig_height = n_rows * 4
    
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(fig_width, fig_height))
    

    if n_plots == 1:
        axes = [axes]
    else:
        axes = axes.flatten()

    for i, col in enumerate(cols):
        ax = axes[i]
        if plot_type == 'boxplot':
            sns.boxplot(data=df, x=target, y=col, ax=ax)
            ax.set_title(f'Distribuição de "{col}" por "{target}"', fontsize=12)
        
        elif plot_type == 'histplot':
            sns.histplot(data=df, x=col, kde=True, ax=ax)
            ax.set_title(f'Histograma de "{col}"', fontsize=12)
            
        elif plot_type == 'countplot':
            sns.countplot(data=df, x=col, ax=ax)
            ax.set_title(f'Contagem de "{col}"', fontsize=12)
            if len(df[col].unique()) > 5:
                ax.tick_params(axis='x', rotation=45)

    for i in range(n_plots, len(axes)):
        axes[i].axis('off')

    plt.tight_layout(pad=3.0) 
    plt.show()

num_cols = df_bank_full.select_dtypes(include='number').columns.tolist()

colunas_para_boxplot = ['age', 'balance', 'duration', 'campaign', 'pdays', 'previous']

plot_grid_dinamico(df=df_bank_full, 
                   cols=colunas_para_boxplot, 
                   plot_type='boxplot', 
                   n_cols=3, 
                   target='y')
```

#### Quais ocupações têm maior taxa de adesão?

```python
adesao_por_profissao = df_bank_full.groupby('job')['y'].value_counts(normalize=True).unstack()
adesao_por_profissao = adesao_por_profissao.sort_values(by='yes', ascending=False)

ax = adesao_por_profissao['yes'].plot(kind='bar', figsize=(14, 7), color='skyblue')

plt.title('Taxa de Adesão por Profissão', fontsize=16)
plt.xlabel('Profissão', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.ylim(0, 0.35)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1%}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.tight_layout()
plt.show()
```

Com base no gráfico da taxa de conversão por ocupação, é possível identificar diferenças bastante marcantes no comportamento dos clientes em relação à adesão à campanha.

As ocupações com maior taxa de adesão são:

Estudantes: apresentaram a maior taxa de conversão entre todas as categorias, indicando uma abertura maior a produtos bancários de investimento. Isso pode refletir um perfil mais otimista, curioso ou com menos compromissos financeiros.

Aposentados: também mostraram alta taxa de adesão. Esse grupo possivelmente busca formas seguras de aplicar suas economias, sendo um público receptivo a produtos de depósito a prazo.

Desempregados: surpreendentemente, aparecem entre os que mais aderem, o que pode indicar uma tendência de tentar obter renda enquanto não obtêm um emprego.

Já ocupações como trabalhadores do setor industrial (blue-collar), empregadas domésticas (housemaid) e empreendedores tiveram as menores taxas de adesão, o que pode refletir uma menor familiaridade com investimentos financeiros, maior resistência ao risco ou falta de interesse nas campanhas propostas.



#### Clientes com nível de escolaridade superior aderem mais que os com nível fundamental ou médio?

```python
ordem_educacional = ['primary', 'secondary', 'tertiary', 'unknown']
df_bank_full['education'] = pd.Categorical(df_bank_full['education'], categories=ordem_educacional, ordered=True)


taxa_adesao_edu = df_bank_full.groupby('education')['y'].value_counts(normalize=True).unstack()['yes']

plt.figure(figsize=(10, 6))
ax = taxa_adesao_edu.plot(kind='bar', color='skyblue', rot=0)

plt.title('Taxa de Adesão por Nível de Escolaridade', fontsize=16)
plt.xlabel('Nível de Escolaridade', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))

for index, value in enumerate(taxa_adesao_edu):
    plt.text(index, value + 0.005, f'{value:.1%}', ha='center', va='bottom')

plt.show()
```

Analisando o gráfico de adesão por nível de escolaridade, observamos que os clientes com educação superior apresentam a maior taxa de conversão, seguidos pelos clientes com escolaridade desconhecida, ensino médio e, por fim, ensino fundamental.

A diferença nas taxas é sutil. Isso sugere que quanto maior o nível de escolaridade, maior a propensão a aderir ao produto oferecido pela campanha. O grupo com educação superior mostra-se mais propenso a aceitar a proposta, possivelmente por entender melhor os benefícios de um investimento financeiro de longo prazo ou por estar em uma situação econômica mais favorável.

Portanto, clientes com nível de escolaridade superior aderem mais do que os com nível fundamental ou médio. Isso indica que segmentar campanhas com foco em pessoas com maior escolaridade pode ser uma estratégia mais eficaz.

#### O estado civil influencia na adesão à campanha?

```python
adesao_por_estadocivil = df_bank_full.groupby('marital')['y'].value_counts(normalize=True).unstack()
adesao_por_estadocivil = adesao_por_estadocivil.sort_values(by='yes', ascending=False)

ax = adesao_por_estadocivil['yes'].plot(kind='bar', figsize=(10, 6), color='lightgreen')

plt.title('Taxa de Adesão por Estado Civil', fontsize=16)
plt.xlabel('Estado Civil', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks(rotation=0)
plt.ylim(0, 0.2)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))

for p in ax.patches:
    ax.annotate(f'{p.get_height():.1%}', (p.get_x() + p.get_width() / 2., p.get_height()),
                ha='center', va='center', xytext=(0, 10), textcoords='offset points')

plt.show()
```

Observando as taxas de adesão de acordo com estado civil, é possível notar que, pessoas que estão solteiras ou já se divorciaram, ou seja, pessoas que atualmente vivem sem um relacionamento, têm maior propensão a realizar investimentos de longo prazo. Isso possivelmente ocorre pois pessoas não comprometidas tendem a ter menos responsabilidades familiares e, consequentemente, mais liberdade para investir ou correr riscos como aplicar em depósitos a prazo.


#### Clientes que possuem financiamento de casa tendem a aceitar ou recusar a oferta?

```python
adesao_por_financiamento = df_bank_full.groupby('housing')['y'].value_counts(normalize=True).unstack()
adesao_por_financiamento = adesao_por_financiamento.sort_values(by='yes', ascending=False)

plt.figure(figsize=(8, 6))

ax = adesao_por_financiamento['yes'].plot(kind='bar', color=['skyblue', 'salmon'], rot=0)

plt.title('Taxa de Adesão por Financiamento Imobiliário', fontsize=16)
plt.xlabel('Possui Financiamento Imobiliário?', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks([0, 1], ['Não', 'Sim'])

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.20)

for index, value in enumerate(adesao_por_financiamento['yes']):
    plt.text(index, value + 0.005, f'{value:.1%}', ha='center', va='bottom')

# Exibe o gráfico
plt.show()
```

Clientes que não possuem um financiamento imobiliário são mais de duas vezes mais propensos a aceitar a oferta do que aqueles que possuem essa dívida.

Isso ocorre porque um financiamento imobiliário representa um compromisso financeiro significativo e de longo prazo, que geralmente consome uma parte considerável da renda mensal do cliente. Como resultado, esses indivíduos tendem a ter menos capital disponível e uma menor flexibilidade financeira para alocar em novos produtos de investimento, como um depósito a prazo.

Dessa forma, a ausência de um financiamento imobiliário é um forte indicador de um cliente com maior potencial de conversão, devendo ser um critério importante na segmentação e priorização do público para futuras campanhas.

#### Ter outros empréstimos pessoais em andamento (loan) afeta negativamente a taxa de adesão?

```python
adesao_por_emprestimo = df_bank_full.groupby('loan')['y'].value_counts(normalize=True).unstack()
adesao_por_emprestimo = adesao_por_emprestimo.sort_values(by='yes', ascending=False)

plt.figure(figsize=(8, 6))

ax = adesao_por_emprestimo['yes'].plot(kind='bar', color=['skyblue', 'salmon'], rot=0)

plt.title('Taxa de Adesão por Empréstimo Pessoal', fontsize=16)
plt.xlabel('Possui Empréstimo Pessoal?', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks([0, 1], ['Não', 'Sim']) 

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.15)

for index, value in enumerate(adesao_por_emprestimo['yes']):
    plt.text(index, value + 0.005, f'{value:.1%}', ha='center', va='bottom')

plt.show()
```

O insight principal aqui é que a taxa de adesão para clientes sem empréstimo pessoal é quase o dobro da taxa para aqueles que possuem um.

Este resultado, somado à análise do financiamento imobiliário, pinta um quadro claro: clientes com compromissos financeiros pré-existentes (seja um empréstimo pessoal ou imobiliário) têm menor capacidade ou interesse em alocar recursos para um depósito a prazo. A existência de uma dívida ativa sinaliza que a prioridade financeira do cliente no momento pode ser a quitação de suas obrigações, e não a acumulação de novas poupanças ou investimentos.

Portanto, a ausência de um empréstimo pessoal é um forte indicador de um cliente com maior potencial de conversão. Este critério, junto com a ausência de financiamento imobiliário, deve ser um dos principais filtros utilizados para qualificar os contatos e direcionar os esforços da campanha de marketing, otimizando o tempo dos operadores e aumentando a chance de sucesso.

#### Clientes com histórico de inadimplência (default) demonstram menor engajamento com a campanha?

```python
adesao_por_default = df_bank_full.groupby('default')['y'].value_counts(normalize=True).unstack()

if 'no' in adesao_por_default.index and 'yes' in adesao_por_default.index:
    adesao_por_default = adesao_por_default.reindex(['no', 'yes'])

plt.figure(figsize=(8, 6))

ax = adesao_por_default['yes'].plot(kind='bar', color=['skyblue', 'salmon'], rot=0)

plt.title('Taxa de Adesão por Histórico de Inadimplência', fontsize=16)
plt.xlabel('Cliente com Histórico de Inadimplência?', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks([0, 1], ['Não', 'Sim'])

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.15) 

for index, value in enumerate(adesao_por_default['yes']):
    plt.text(index, value + 0.005, f'{value:.1%}', ha='center', va='bottom')

plt.show()
```

O principal insight é que a taxa de adesão para clientes sem histórico de default é consideravelmente maior, quase o dobro daquela para clientes que já foram inadimplentes.

Um histórico de inadimplência é um forte sinal de dificuldades financeiras passadas, o que pode indicar uma instabilidade financeira presente ou, no mínimo, uma abordagem mais cautelosa em relação a novos produtos financeiros. É natural que clientes com esse perfil sejam menos propensos a alocar fundos em um depósito a prazo, que imobiliza o dinheiro por um determinado período.

Este resultado, somado às análises de empréstimo pessoal e financiamento imobiliário, reforça uma conclusão central do projeto: a saúde financeira do cliente é um dos pilares mais importantes para prever o sucesso da campanha. Clientes com um perfil financeiro estável, sem dívidas ativas e sem histórico de inadimplência, são o público-alvo ideal e com maior probabilidade de conversão

#### A taxa de sucesso é maior quando o contato é feito por celular (cellular) ou telefone fixo (telephone)?

```python
adesao_por_contato = df_bank_full.groupby('contact')['y'].value_counts(normalize=True).unstack()

adesao_por_contato = adesao_por_contato.sort_values(by='yes', ascending=False)

plt.figure(figsize=(10, 6))

ax = adesao_por_contato['yes'].plot(kind='bar', color=['skyblue', 'salmon', 'lightgreen'], rot=0)

plt.title('Taxa de Adesão por Tipo de Contato', fontsize=16)
plt.xlabel('Tipo de Contato', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.20) 

for index, value in enumerate(adesao_por_contato['yes']):
    plt.text(index, value + 0.005, f'{value:.1%}', ha='center', va='bottom')

plt.show()
```

O insight principal, baseado neste gráfico, é que, embora ambos os canais de contato (celular e telefone) sejam viáveis, o celular apresenta uma eficácia ligeiramente superior na conversão de clientes.

A diferença entre os dois canais principais não é tão drástica quanto em análises anteriores, mas ainda assim favorece o contato através do celular. Isso sugere que, ao otimizar a campanha, dar uma leve prioridade para os contatos via celular pode aumentar a taxa de sucesso geral.

#### Qual mês da campanha apresenta maior adesão? Existe um período mais favorável para o contato?

```python
ordem_meses = ['mar', 'apr', 'may', 'jun', 'jul', 'aug', 'sep', 'oct', 'nov', 'dec']
df_bank_full['month'] = pd.Categorical(df_bank_full['month'], categories=ordem_meses, ordered=True)

adesao_por_mes = df_bank_full.groupby('month')['y'].value_counts(normalize=True).unstack()

plt.figure(figsize=(12, 7))

ax = adesao_por_mes['yes'].plot(kind='bar', color='skyblue', rot=0)

plt.title('Taxa de Adesão por Mês do Contato', fontsize=16)
plt.xlabel('Mês', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.6) 

for index, value in enumerate(adesao_por_mes['yes']):
    plt.text(index, value + 0.015, f'{value:.1%}', ha='center', va='bottom', fontsize=10)

plt.tight_layout()
plt.show()
```

A análise do mês em que o contato foi realizado demonstra que a eficácia da campanha varia muito ao longo do ano. Existem períodos de alta receptividade por parte dos clientes e outros em que o esforço de marketing gera um retorno muito baixo.

Períodos de Alta Performance (Picos de Sucesso):

O gráfico revela dois picos de sucesso evidentes. O primeiro ocorre no início da primavera (no hemisfério norte), com Março (mar) sendo o mês de melhor desempenho absoluto, atingindo uma taxa de adesão de 52,0%.

O segundo período de alta acontece no outono e no final do ano. Setembro (sep), Outubro (oct) e Dezembro (dec) apresentam taxas bastante altas, com 46,5%, 43,8% e 47,8%, respectivamente.

Período de Baixa Performance (Vale):

O mês de Maio (may) se mostra o pior momento para a campanha, com a taxa de adesão despencando para apenas 6,7%. Os meses de verão no geral (Junho, Julho e Agosto) também apresentam um desempenho modesto.

Conclusão e Recomendações Estratégicas:

A conclusão é que o timing da campanha é fundamental. O esforço de contatar um cliente em Março tem uma chance de sucesso quase 8 vezes maior do que contatá-lo em Maio.

Por quê? As razões podem ser diversas: estes picos podem coincidir com períodos em que os clientes recebem bônus de fim de ano, restituição de impostos, ou simplesmente estão em um momento de maior planejamento financeiro. O baixo desempenho no verão europeu pode estar relacionado a períodos de férias, quando as pessoas estão menos focadas em decisões financeiras.

Recomendação Estratégica: A alocação de recursos de marketing deve ser completamente redesenhada com base nesses dados. A recomendação é concentrar massivamente os esforços e o orçamento nos períodos de pico (Março e de Setembro a Dezembro) e reduzir drasticamente ou até pausar a campanha durante os meses de baixa, especialmente em Maio. Isso irá otimizar o investimento e aumentar exponencialmente o número de conversões com o mesmo ou menor esforço.

#### Clientes que participaram de campanhas anteriores e não aderiram (poutcome = failure) ainda têm chance de conversão?

```python
adesao_por_resultado = df_bank_full.groupby('poutcome')['y'].value_counts(normalize=True).unstack()

adesao_por_resultado = adesao_por_resultado.sort_values(by='yes', ascending=False)

plt.figure(figsize=(10, 7))

ax = adesao_por_resultado['yes'].plot(kind='bar', color=['gold', 'skyblue', 'lightgreen', 'salmon'], rot=0)

plt.title('Taxa de Adesão por Resultado da Campanha Anterior', fontsize=16)
plt.xlabel('Resultado da Campanha Anterior', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.75)
for index, value in enumerate(adesao_por_resultado['yes']):
    plt.text(index, value + 0.015, f'{value:.1%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

1 - Clientes do grupo failure ainda são valiosos.
O gráfico mostra que o grupo failure teve uma taxa de adesão de 12,9% na campanha atual.Essa taxa de conversão não é desprezível. Na verdade, ela é mais alta do que a de vários outros segmentos que analisamos (como clientes com empréstimos, com histórico de inadimplência ou de certas profissões). Isso significa que desistir desses clientes seria um erro. Eles já conhecem a oferta e, com uma nova abordagem ou no momento certo, ainda podem converter.

2 - Sucesso passado é o melhor previsor de sucesso futuro.O grupo de clientes que teve sucesso (success) em campanhas passadas é, de longe, o mais propenso a aderir novamente, com uma taxa de conversão extraordinária de 64,7%. Este é o segmento de clientes mais valioso e de maior prioridade. A probabilidade de conversão aqui é mais de 4 vezes maior do que a do segundo melhor grupo (other). A estratégia para esses clientes deve ser de retenção e relacionamento, talvez com ofertas exclusivas, pois eles representam o retorno mais garantido.

3 - A segmentação por poutcome é a estratégia mais eficaz.O gráfico nos permite criar uma hierarquia clara de prioridade para os clientes: Prioridade Máxima: o grupo success (64,7%). Prioridade Média: os grupos other (16,7%) e failure (12,9%). São clientes que já foram engajados e apresentam um potencial de conversão sólido. Prioridade Baixa: O grupo unknown (9,2%). Este é o maior grupo em volume e representa o mercado geral, que nunca foi convertido antes. A comunicação aqui é de prospecção, sabendo que a taxa de sucesso será naturalmente menor.

Conclusão Estratégica:
A campanha não deve tratar todos os clientes da mesma forma. A implementação de uma estratégia de marketing por camadas, baseada no histórico do cliente (poutcome), é a recomendação mais forte desta análise. Direcionar os melhores recursos para o grupo success e criar campanhas de reengajamento para os grupos failure.

#### Com base nas combinações de perfil (educação superior + sem empréstimos + contato por celular), quais segmentos possuem a maior taxa de adesão?

```python
filtro_perfil = (df_bank_full['education'] == 'tertiary') & \
                (df_bank_full['housing'] == 'no') & \
                (df_bank_full['loan'] == 'no') & \
                (df_bank_full['contact'] == 'cellular')

df_segmento_ideal = df_bank_full[filtro_perfil]

adesao_no_segmento = df_segmento_ideal.groupby('job')['y'].value_counts(normalize=True).unstack().fillna(0)

adesao_no_segmento = adesao_no_segmento.sort_values(by='yes', ascending=False)


plt.figure(figsize=(12, 7))
ax = adesao_no_segmento['yes'].plot(kind='bar', color='purple', rot=45)

plt.title('Taxa de Adesão por Profissão no Segmento de Alto Potencial', fontsize=16)
plt.xlabel('Profissão', fontsize=12)
plt.ylabel('Taxa de Adesão (%)', fontsize=12)
plt.xticks(ha='right') 

ax.yaxis.set_major_formatter(plt.FuncFormatter('{:.0%}'.format))
plt.ylim(0, 0.5)

for index, value in enumerate(adesao_no_segmento['yes']):
    if value > 0: # Adiciona rótulo apenas se a barra existir
        plt.text(index, value + 0.01, f'{value:.1%}', ha='center', va='bottom')

plt.tight_layout()
plt.show()
```

Insights finais
Este gráfico analisa a taxa de adesão por profissão, mas apenas dentro do grupo de clientes que já consideramos de alto potencial: aqueles com educação superior, sem empréstimos (pessoal ou imobiliário) e que foram contatados por celular.

Principal segmento:
Mesmo dentro de um grupo já altamente qualificado, duas profissões se destacam. Estudantes (student) e Aposentados (retired) atingem taxas de adesão excelentes de 43,1% e 37,8%, respectivamente. Isso é um insight poderoso. Significa que uma ligação para um estudante ou aposentado que se encaixa nesse perfil tem uma chance de sucesso entre 1 em 2 e 1 em 3. Para uma campanha de marketing, este é um resultado extraordinário e representa o melhor da base de clientes.

O segundo melhor segmento: encontramos outro grupo com um desempenho muito bom: Desempregados (unemployed), com 26,5% de adesão. Este é um achado interessante, sugerindo que pessoas nessa condição podem estar buscando ativamente por segurança financeira.

Profissionais de gestão (management) e autônomos (self-employed) também convertem a uma taxa considerável, próxima de 20%.

A Validação do Filtro:

O insight mais importante aqui é a validação da nossa estratégia de segmentação. A profissão com o pior desempenho dentro deste grupo de elite (blue-collar com 10,9%) ainda possui uma taxa de adesão próxima da média geral de todo o dataset (que era de 11,7%). Isso prova que filtrar por educação, ausência de empréstimos e contato por celular eleva o potencial de todos os clientes, tornando todo o esforço de marketing dentro deste segmento muito mais eficiente.

Conclusão:

Prioridade Máxima: Estudantes e Aposentados.

Prioridade Alta: Desempregados, Gestores e Autônomos.

Dentro deste grupo, estudantes e aposentados representam o melhor segmento, com taxas de adesão bastante altas. Portanto, a recomendação final é a adoção de uma abordagem de forte segmentação, priorizando os recursos de marketing para este perfil ideal nos períodos mais favoráveis, a fim de maximizar o retorno sobre o investimento e a eficiência da campanha.
